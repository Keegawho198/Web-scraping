```python
from splinter import Browser
from bs4 import BeautifulSoup
import pandas as pd
import time

# Specify the URL
url = 'https://www.nba.com/stats/players/traditional?PerMode=Totals&sort=PTS&dir=-1&Season=2018-19'

# Setup Splinter browser with Chrome
with Browser('chrome') as browser:
    # Open the URL
    browser.visit(url)
    
    # Give the page time to load all content
    time.sleep(5)
    
    # Parse the page with BeautifulSoup
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    
    # Find the table
    table = soup.find('table', {'class': 'Crom_table__p1iZz'})
    
    if table:
        print("Table found!")
        
        # Extract table headers
        headers = [header.text.strip() for header in table.find_all('th')]
        print("Headers count:", len(headers))
        print("Headers:", headers)
        
        # Extract table rows
        rows = []
        for row in table.find_all('tr')[1:]:  # Skip header row
            cells = row.find_all('td')
            row_data = [cell.text.strip() for cell in cells]
            rows.append(row_data)
        
        # Check if rows were populated
        print("Row count:", len(rows))
        if rows:
            print("First row length:", len(rows[0]))  # Print the length of the first row
            
            # Slice headers if there are more headers than row columns
            if len(headers) > len(rows[0]):
                headers = headers[:len(rows[0])]
                print("Adjusted headers:", headers)
            
            # Create DataFrame
            df = pd.DataFrame(rows, columns=headers)
            print(df)  # Print the DataFrame
        else:
            print("No rows found in the table.")
    else:
        print("Table not found.")
```
